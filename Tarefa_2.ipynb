##### Divida seus dados em conjuntos de treinamento / teste (90/10)

```python
X_dados, X_test, y_Classes, y_test = train_test_split(df.data, df.target, test_size=0.10)
```

##### Aplique as etapas de pré-processamento sobre seus dados que você acha necessárias (dimensionamento de recursos, normalização)

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
```

##### Treine com validação cruzada dois métodos SL diferentes na tarefa de classificação para a classificação de 10 dígitos

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score

from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
KFold_Strat = StratifiedKFold(n_splits=5)
```

```python
resultados_acuracia = []
resultados_precisao = []

for indice_train, indice_valid in KFold_Strat.split(X_dados, y_Classes):
    #print(" train ", X_train[train])
    #print("Validação",  y_train[y_valid])
    print("----------")
    
    #classificador = SVC(kernel='linear', C=2, random_state=42)
    classificador = SVC(kernel = 'rbf', random_state = 2, C = 3.0)
    
    classificador.fit(X_dados[indice_train], y_Classes[indice_train])
    
    previsoes = classificador.predict(X_dados[indice_valid])
    
    acuracia = accuracy_score(y_Classes[indice_valid], previsoes)
    precisao =  precision_score(y_Classes[indice_valid], previsoes, average='macro')
    resultados_acuracia.append(acuracia)
    resultados_precisao.append(precisao)
    
    
    print(classification_report(y_Classes[indice_valid], previsoes))
    
    
    print("acuracia  ", acuracia)
    print("Precisão ", precisao)
resultados_acuracia = np.asarray(resultados_acuracia)
resultados_precisao = np.asarray(resultados_precisao)
print("Media Da acuracia" , resultados_acuracia.mean())
print("Media Da precisao" , resultados_precisao.mean())
    
 
```
